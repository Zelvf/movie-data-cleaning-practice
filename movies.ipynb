```python
#import necessary libraries
import re
import pandas as pd
from datetime import datetime
```

```python
#READ and print out the data

movies = pd.read_csv('movies.csv')
movies
```

```python
#split the year column to start and end year

def extract_years(year_val):
    try:
        years = re.findall(r'\d{4}', str(year_val))  # find all 4-digit numbers
        
        if not years:
            return pd.Series([None, None])
        
        start = int(years[0])
        
        # Check if there is a second year in the string
        if len(years) > 1:
            end = int(years[1])
        else:
            # Check if original string ends with a dash (meaning ongoing series)
            if str(year_val).strip().endswith('– )') or str(year_val).strip().endswith(' )'):
                end = datetime.now().year # ongoing, so end year is unknown
            else:
                # Single year case
                end = start
        
        return pd.Series([start, end])
    
    except:
        return pd.Series([None, None])

# Apply the function and drop the original YEAR column
movies[['start_year', 'end_year']] = movies['YEAR'].apply(extract_years)
movies.drop(columns=['YEAR'], inplace=True)
```

```python
#define a function that removes the \n woth a space
def remove_newlines (rem):
    return rem.applymap(lambda x: x.replace('\n', ' ') if isinstance(x, str) else x)

movies = remove_newlines(movies)
```

```python
#check for nans values
print(movies.isna().sum())
```

```python
#delete the gross columns since it mostly contains nans
movies.drop(columns=['Gross'], inplace=True)
```

```python
#drops rows without a plot
# Remove rows where 'ONE-LINE' equals 'Add a Plot' after stripping whitespace
movies = movies[~movies['ONE-LINE'].str.strip().eq("Add a Plot")]
```

```python
#counts number of duplicate movie titles
duplicate = movies['MOVIES'].duplicated().sum()
print(f"Number of duplicate movie titles: {duplicate}")
```

```python
# First, identify the titles that are duplicated
duplicated_titles = movies[movies['MOVIES'].duplicated(keep=False)]['MOVIES'].unique()
print(f"Found {len(duplicated_titles)} unique movie titles that have duplicates.")

#  show only rows with these duplicated titles
movies_with_duplicate_titles = movies[movies['MOVIES'].isin(duplicated_titles)].sort_values(by=['MOVIES', 'start_year'])

# Display a sample of these duplicate titles.
print("\n--- Sample of Rows with Duplicated Titles ---")
# Display the first 20 rows of these duplicates, and specifically relevant columns
print(movies_with_duplicate_titles[['MOVIES', 'start_year', 'end_year', 'RATING', 'VOTES', 'RunTime']].head(20))
```
